```python
import streamlit as st
import pandas as pd

# Load your data directly from the file path
df = pd.read_excel("caprae.xlsx")

# The root sequence is assumed to be the first row
root_sequence = df.iloc[0, 2:]

# Sidebar for selecting isolates or groups
st.sidebar.title("DNA Isolate Comparator")
isolate_options = df['Unnamed: 0'].unique()  # Extract the isolate names
selected_isolates = st.sidebar.multiselect("Select Isolates", isolate_options, default=[isolate_options[1]])

# Marker selection (small sequence of DNA)
available_markers = root_sequence.index
selected_markers = st.sidebar.multiselect("Select Markers (up to 5)", available_markers, default=available_markers[:5])

# Filter and compare isolates
def compare_isolates(isolate_names, markers=None):
    selected_isolates = df[df['Unnamed: 0'].isin(isolate_names)].iloc[:, 2:]
    if markers is not None and len(markers) > 0:
        selected_isolates = selected_isolates[markers]
        root_filtered = root_sequence[markers]
    else:
        root_filtered = root_sequence

    differences = selected_isolates != root_filtered
    row_diff_count = differences.sum(axis=0)  # Count differences by row (markers)
    col_diff_count = differences.sum(axis=1)  # Count differences by column (isolates)

    return differences, row_diff_count, col_diff_count

# Run comparison and display results
if selected_isolates:
    differences, row_diff_count, col_diff_count = compare_isolates(selected_isolates, selected_markers)

    # Display the comparison table with differences
    st.write("Comparison Table:")
    st.dataframe(differences)

    # Bar chart showing differences by marker (row-wise frequency)
    st.write("Differences by Marker (Row-wise):")
    st.bar_chart(row_diff_count)

    # Bar chart showing differences by isolate (column-wise frequency)
    st.write("Differences by Isolate (Column-wise):")
    st.bar_chart(col_diff_count)

# Use Streamlit to run: `streamlit run your_script.py`
```
